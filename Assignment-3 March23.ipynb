Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

```python
Ans-Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Seaborn is particularly useful for visualizing complex datasets and exploring relationships within data. Here are five types of plots that you can create using Seaborn along with their typical use cases:

Histograms (sns.histplot):

Use: Histograms are used to visualize the distribution of a single numeric variable. They show the frequency or density of data within specified bins or intervals.
Scatter Plots (sns.scatterplot):

Use: Scatter plots are used to visualize the relationship between two numeric variables. They help identify patterns, trends, clusters, and outliers in data.
Box Plots (sns.boxplot):

Use: Box plots (box-and-whisker plots) are useful for visualizing the distribution of a numeric variable or comparing distributions between different categories. They display summary statistics, such as the median, quartiles, and potential outliers.
Bar Plots (sns.barplot):

Use: Bar plots are used to visualize the relationship between a categorical variable and a numeric variable. They show the mean or other summary statistics of the numeric variable within each category.
Heatmaps (sns.heatmap):

Use: Heatmaps are used to display the relationships between two categorical variables or between one categorical and one numeric variable. They represent data as a grid of colored squares, with each color indicating the strength or frequency of the relationship.
These are just a few examples of the types of plots you can create with Seaborn. Seaborn offers a wide range of plot types and customization options, making it a valuable tool for data exploration, data analysis, and data presentation.





Regenerate
```

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot for "signal" vs. "timepoint" for different "events" and "regions"
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region")

# Add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("Line Plot of Signal vs. Timepoint for Different Events and Regions")

# Show the legend
plt.legend(title="Event")

# Display the plot
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Create the first box plot for 'age' vs. 'pclass'
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)  # Create a subplot (1 row, 2 columns), select the first subplot
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age vs. Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Create the second box plot for 'fare' vs. 'pclass'
plt.subplot(1, 2, 2)  # Select the second subplot
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare vs. Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

# Adjust subplot spacing
plt.tight_layout()

# Show the plots
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Create a histogram for 'price' with 'cut' as the hue
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", bins=30, multiple="stack", palette="husl")

# Add labels and a title
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")

# Show the legend
plt.legend(title="Cut")

# Display the plot
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create a pair plot with hue for the "species" column
sns.pairplot(data=iris_data, hue="species", palette="husl")

# Add a title
plt.suptitle("Pair Plot of Iris Dataset by Species")

# Show the plot
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d", linewidths=0.5)

# Add labels and a title
plt.xlabel("Year")
plt.ylabel("Month")
plt.title("Passenger Counts Heatmap (Flights Dataset)")

# Show the plot
plt.show()
```
