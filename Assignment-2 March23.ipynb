Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

```python
Ans-Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in 2D and 3D. It provides a flexible and comprehensive toolkit for creating a wide range of charts, graphs, and plots, making it an essential tool for data visualization and data exploration tasks. Matplotlib is widely used in various fields, including data analysis, scientific research, machine learning, and more.

Matplotlib's key features and capabilities include:

Customizability: Matplotlib allows users to customize every aspect of their plots, including colors, fonts, labels, axes, and more, to create publication-quality visualizations.

Wide Range of Plot Types: Matplotlib supports a variety of plot types, making it suitable for diverse visualization needs.

Five common types of plots that can be created using the Pyplot module of Matplotlib are:

Line Plot: Line plots are used to visualize data points as connected line segments. They are suitable for showing trends and changes over time or along a continuous variable.

Scatter Plot: Scatter plots display individual data points as markers on a two-dimensional plane. They are useful for visualizing relationships between two variables or identifying clusters in data.

Bar Plot: Bar plots, including bar charts and histograms, are used to display data in rectangular bars. Bar charts are suitable for comparing categorical data, while histograms show the distribution of continuous data.

Pie Chart: Pie charts represent data as sectors of a circle, with each sector's size proportional to the quantity it represents. They are used to show the composition of a whole.

Box Plot: Box plots, also known as box-and-whisker plots, provide a summary of the distribution of a dataset. They display the median, quartiles, and potential outliers, making them useful for understanding the spread of data.

```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, label='Data Points', color='blue', marker='o')

# Add title and labels
plt.title('Scatter Plot of x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Display the legend (optional)
plt.legend()

# Show the plot
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
plt.figure(figsize=(12, 6))

# Plot line 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='Line 1')
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot line 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2, label='Line 2', color='orange')
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot line 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3, label='Line 3', color='green')
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Plot line 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4, label='Line 4', color='red')
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Adjust subplot spacing
plt.tight_layout()

# Show the subplots
plt.show()
```
