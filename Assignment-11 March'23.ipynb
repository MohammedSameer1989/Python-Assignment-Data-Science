Q2: Differentiate between one-tailed and two-tailed tests

```python
One-Tailed Test and Two-Tailed Test are types of hypothesis tests used in statistics. They differ in how they assess the relationship between the sample data and the null hypothesis. Here's the key difference between one-tailed and two-tailed tests:

One-Tailed Test:

A one-tailed test, also known as a one-sided test, is used to test a specific directional hypothesis. It is designed to determine whether the sample data is significantly greater than or less than a particular value or a hypothesized population parameter.
In a one-tailed test, you have an alternative hypothesis that specifies the direction of the effect (greater than or less than), and the null hypothesis represents the absence of that effect.
The critical region, where extreme values of the test statistic lead to rejection of the null hypothesis, is located in only one tail of the probability distribution.
One-tailed tests are typically used when you have a clear expectation or hypothesis about the direction of the effect you are investigating.
Two-Tailed Test:

A two-tailed test, also known as a two-sided test, is used to test a non-directional hypothesis. It is designed to determine whether the sample data is significantly different from a particular value or a hypothesized population parameter.
In a two-tailed test, the alternative hypothesis typically states that there is a difference or an effect, but it doesn't specify the direction (greater than or less than).
The critical region for a two-tailed test is divided into two equal tails, with extreme values in either tail leading to rejection of the null hypothesis.
Two-tailed tests are used when you are interested in whether the sample data is significantly different from the null hypothesis value but you don't have a prior expectation about the direction of the difference.
Summary:

Use a one-tailed test when you have a specific directional hypothesis and want to test whether the sample data is significantly greater than or less than a particular value or parameter.
Use a two-tailed test when you have a non-directional hypothesis and want to test whether the sample data is significantly different from a particular value or parameter without specifying the direction of the difference.
In both cases, the choice of whether to use a one-tailed or two-tailed test should be based on the research question, the nature of the hypothesis, and the specific context of the study. It's important to make this choice carefully to ensure that the statistical test aligns with the research objectives.
```

```python
Type 1 and Type 2 errors are two different kinds of errors that can occur in hypothesis testing:

Type 1 Error (False Positive):

A Type 1 error occurs when you reject a null hypothesis that is actually true. In other words, it is the incorrect rejection of a null hypothesis when there is no real effect or difference in the population.
The probability of making a Type 1 error is denoted by 
�
α and is also called the significance level. It represents the probability of observing a significant result when there is no true effect.
Type 1 errors are more likely to occur when you set a high significance level (e.g., 
�
=
0.10
α=0.10).
Example Scenario for Type 1 Error:
Imagine you are testing a new drug for a disease. The null hypothesis (
�
0
H 
0
​
 ) states that the drug has no effect, and the alternative hypothesis (
�
1
H 
1
​
 ) states that the drug is effective. If, in reality, the drug is ineffective, but your study (due to random variation) shows a significant improvement in the treated group and you reject the null hypothesis, you commit a Type 1 error.

Type 2 Error (False Negative):

A Type 2 error occurs when you fail to reject a null hypothesis that is actually false. In other words, it is the incorrect retention of a null hypothesis when there is a real effect or difference in the population.
The probability of making a Type 2 error is denoted by 
�
β. It represents the probability of not observing a significant result when there is a true effect.
Type 2 errors are more likely to occur when the sample size is small, the effect size is small, or the significance level is set low (e.g., 
�
=
0.01
α=0.01).
Example Scenario for Type 2 Error:
Suppose you are conducting a quality control test for a manufacturing process. The null hypothesis (
�
0
H 
0
​
 ) states that the process is in control and producing items within specification. The alternative hypothesis (
�
1
H 
1
​
 ) states that the process is out of control, and items are defective. If, in reality, the process is out of control and producing defective items, but your study (due to random variation) fails to detect the problem, you commit a Type 2 error.

In summary:

Type 1 error involves erroneously rejecting a true null hypothesis.
Type 2 error involves erroneously retaining a false null hypothesis.
The balance between Type 1 and Type 2 errors is controlled by factors like the significance level, sample size, effect size, and variability in the data.
Researchers often aim to minimize the risk of both types of errors by choosing appropriate significance levels and conducting power analysis to determine suitable sample sizes.
```
