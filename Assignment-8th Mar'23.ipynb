Q2. What are the types of Probability distribution?

```python
There are several types of probability distributions in statistics, each describing how the probabilities of different outcomes or values are distributed for a random variable. The choice of distribution depends on the characteristics of the data and the underlying processes being modeled. Some common types of probability distributions include:

Discrete Probability Distributions:

Bernoulli Distribution: Models a binary outcome with two possible values (usually 0 and 1), such as success/failure or heads/tails in a coin toss.
Binomial Distribution: Describes the number of successes (usually denoted as "k") in a fixed number of independent Bernoulli trials.
Poisson Distribution: Models the number of events occurring in a fixed interval of time or space, given a known average rate.
Continuous Probability Distributions:

Normal Distribution (Gaussian Distribution): Often referred to as the bell curve, it is characterized by its symmetric, bell-shaped curve and is widely used to model a variety of natural phenomena.
Uniform Distribution: All values within a specified interval have equal probabilities.
Exponential Distribution: Models the time between events in a Poisson process and is often used to describe the distribution of waiting times.
Log-Normal Distribution: The logarithm of a random variable follows a normal distribution, which makes it useful for modeling data that is inherently positive and right-skewed.
Continuous Probability Distributions with Shapes Depending on Parameters:

Gamma Distribution: Generalizes the exponential distribution and is used for modeling the waiting time until a Poisson process reaches a certain number of events.
Beta Distribution: Often used to model probabilities or proportions and has a flexible shape depending on its parameters.
Chi-Squared Distribution: Arises in the context of hypothesis testing and confidence interval estimation.
Student's t-Distribution: Used for inference about population means when the sample size is small and population variance is unknown.
Multinomial Distribution: Generalizes the binomial distribution to more than two categories or outcomes.

Hypergeometric Distribution: Models the number of successes in a finite population without replacement.

Geometric Distribution: Models the number of trials needed to achieve the first success in a sequence of independent Bernoulli trials.

Weibull Distribution: Often used in reliability engineering and survival analysis to model time-to-failure data.

Cauchy Distribution: Known for its heavy tails and lack of finite moments, making it suitable for modeling extreme values.

Pareto Distribution: Used to describe distributions with heavy tails, such as income and wealth distributions.
```

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
import math

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point.

    Args:
    x (float): The point at which to calculate the PDF.
    mean (float): The mean (average) of the normal distribution.
    std_dev (float): The standard deviation (a measure of spread) of the normal distribution.

    Returns:
    float: The PDF value at the given point.
    """
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -((x - mean) ** 2) / (2 * (std_dev ** 2))
    pdf = coefficient * math.exp(exponent)
    return pdf

# Example usage:
mean = 0.0
std_dev = 1.0
x_value = 1.5
pdf_at_x = normal_pdf(x_value, mean, std_dev)
print(f'PDF at x={x_value}: {pdf_at_x}')
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
import math

def poisson_cdf(k, mean):
    """
    Calculate the Cumulative Distribution Function (CDF) of a Poisson distribution at a given point.

    Args:
    k (int): The point at which to calculate the CDF.
    mean (float): The mean of the Poisson distribution.

    Returns:
    float: The CDF value at the given point.
    """
    if k < 0:
        return 0.0  # CDF is 0 for negative values of k
    cdf = 0.0
    for i in range(k + 1):
        cdf += (math.exp(-mean) * (mean ** i)) / math.factorial(i)
    return cdf

# Example usage:
mean = 3.0
k_value = 2
cdf_at_k = poisson_cdf(k_value, mean)
print(f'CDF at k={k_value}: {cdf_at_k}')
```

Q7. How Binomial distribution different from Poisson distribution?

```python
The Binomial distribution and the Poisson distribution are both probability distributions used in statistics, but they have distinct differences in terms of their characteristics and the types of phenomena they are used to model. Here are the key differences between the two distributions:

Nature of Outcomes:

Binomial Distribution: The Binomial distribution is used to model discrete random variables that represent the number of successes (usually denoted as "k") in a fixed number of independent and identical trials or experiments (usually denoted as "n"). Each trial has only two possible outcomes: success or failure.
Poisson Distribution: The Poisson distribution, on the other hand, is used to model the number of events (often denoted as "k") occurring in a fixed interval of time or space. It is used for rare events or events that occur randomly over time or space.
Number of Trials:

Binomial Distribution: The number of trials (n) is fixed and known in advance. Each trial is independent of the others, and the probability of success (p) remains constant across all trials.
Poisson Distribution: There is no fixed number of trials. The Poisson distribution deals with events that can occur at any time and with a rate parameter (λ) that represents the average number of events in the given interval.
Probability of Success:

Binomial Distribution: The probability of success (p) is constant and the same for each trial. The probability of failure (q) is also constant and complementary to p (i.e., q = 1 - p).
Poisson Distribution: The Poisson distribution does not involve success or failure probabilities like the Binomial distribution. Instead, it is characterized by a single parameter (λ) that represents the average rate of event occurrences.
Outcome Space:

Binomial Distribution: The random variable in a Binomial distribution can take on integer values from 0 to n (inclusive), representing the number of successes in n trials.
Poisson Distribution: The random variable in a Poisson distribution can take on integer values starting from 0 and extending to positive infinity, representing the count of events in the given interval.
Typical Applications:

Binomial Distribution: It is commonly used to model scenarios such as the number of heads in a series of coin flips, the number of defective items in a batch of production, or the number of successful email responses in a marketing campaign.
Poisson Distribution: It is used for modeling rare events such as the number of customer arrivals at a store in an hour, the number of car accidents at a specific intersection in a day, or the number of typos in a book.
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Set the parameters
mean = 5  # Mean of the Poisson distribution
sample_size = 1000  # Sample size

# Generate a random sample from the Poisson distribution
random_sample = np.random.poisson(mean, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=1)  # Use ddof=1 for sample variance

# Print the results
print(f'Sample Mean: {sample_mean}')
print(f'Sample Variance: {sample_variance}')
```
